```python
import pandas as pd
import numpy as np
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
df1 = pd.read_csv(url)
df1.columns = df1.columns.str.lower()
df1.columns = df1.columns.str.replace(" ", "_")
df1.rename(columns= {'st': 'state'}, inplace=True)
df1
```

```python
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv'
df2 = pd.read_csv(url)
df2.columns = df2.columns.str.lower()
df2.columns = df2.columns.str.replace(" ", "_")
df2.rename(columns= {'st': 'state'}, inplace=True)
df2
```

```python
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv'
df3 = pd.read_csv(url)
df3.columns = df3.columns.str.lower()
df3.columns = df3.columns.str.replace(" ", "_")
df3.rename(columns= {'st': 'state'}, inplace=True)
df3
```

```python
df = pd.concat([df1, df2, df3])
df
```

```python
#df1.set_index('customer', inplace=True)
#df2.set_index('customer', inplace=True)
#df3.set_index('customer', inplace=True)
#df1.join([df2,df3],how='outer')
#merged_df = pd.merge(df1, df2, on='customer', how='outer')
#merged_df
```

```python
# check duplicated raws and duplicated customer IDs
print(f'number of duplicated rows in the dataframe:  {df.duplicated().sum()} ' )
print(f"number of duplicated rows in the customer ID:  {df.duplicated(subset=['customer']).sum()} " )
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
print(f'number of duplicated rows in the dataframe:  {df.duplicated().sum()} ' )
```

```python
print(f"number of duplicated rows in the customer ID:  {df.duplicated(subset=['customer']).sum()} " )
```

```python
#Cleaning gender
print(df['gender'].value_counts())
gender_mapping = {'F': 'F', 'M': 'M','Male': 'M', 'female': 'F', 'Femal': 'F'}
df['gender'] = df['gender'].map(gender_mapping)
print(df['gender'].value_counts())
```

```python
#Cleaning state
print(df['state'].value_counts())
df['state'] = df['state'].str.replace("WA", "Washington")
df['state'] = df['state'].str.replace("AZ", "Arizona")
df['state'] = df['state'].str.replace(r"\bCali\b", "California")
print(df['state'].value_counts())
```

```python
#Cleaning education
print(df['education'].value_counts())
df['education'] = df['education'].str.replace("Bachelors", "Bachelor")
print(df['education'].value_counts())
```

```python
#Cleaning Customer life time
df['customer_lifetime_value'] = df['customer_lifetime_value'].str.rstrip('%').astype('float')
df
```

```python
#Cleaning vehicle class
replacements = {
    'Sports Car': 'Luxury',
    'Luxury SUV': 'Luxury',
    'Luxury Car': 'Luxury'
}

print(df['vehicle_class'].value_counts())
df['vehicle_class'] = df['vehicle_class'].replace(replacements)
print(df['vehicle_class'].value_counts())
```

```python
df['# complaints'] = df['number_of_open_complaints'].str.split('/').str[1].astype('float')
df
```

```python
df.info()
```

```python
print(f"number of duplicated rows in the customer ID:  {df.duplicated(subset=['customer']).sum()} " )
duplicates = df[df.duplicated('customer', keep=False)]
duplicates.sort_values(by=['customer'])
```

```python
merged_df = df.groupby('customer').apply(lambda x: x.fillna(method='ffill').fillna(method='bfill'))
```

```python
merged_df
```

```python
merged_df.drop_duplicates(subset=['customer'], keep='first', inplace=True)
```

```python
merged_df
```

```python
print(f"number of duplicated rows in the customer ID:  {merged_df.duplicated(subset=['customer']).sum()} " )
```
